# Create a new table in a database

```python
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv('../application/.env'))

```

```python
from sqlalchemy import create_engine, Column, Integer, Text, Table, MetaData
from sqlalchemy.ext.declarative import declarative_base
import os
from sqlalchemy import create_engine, text

# Retrieve the environment variables for the connection
host = os.getenv("PG_VECTOR_HOST")
user = os.getenv("PG_VECTOR_USER")
password = os.getenv("PG_VECTOR_PASSWORD")
database = os.getenv("PGDATABASE5")  # question_answer

# Create the connection string
CONNECTION_STRING = f"postgresql+psycopg://{user}:{password}@{host}:5432/{database}"
print(CONNECTION_STRING)

# Create a base class using declarative_base
Base = declarative_base()

# Define your table as a model class
class FAQ(Base):
    __tablename__ = 'question_answer'  # Table name in the database

    id = Column(Integer, primary_key=True, autoincrement=True)  # Auto-incrementing primary key
    question = Column(Text, nullable=False)  # Question column of type text
    answer = Column(Text, nullable=False)  # Answer column of type text
    bot_type = Column(Text, nullable=False)  # Bot type column of type text

# Create an engine that connects to the PostgreSQL database
engine = create_engine(CONNECTION_STRING)

# Create the table in the database
Base.metadata.create_all(engine)
```

# Inspect a table in a database

```python
from sqlalchemy import create_engine, inspect


# Inspect the database to see if the table exists
inspector = inspect(engine)
tables = inspector.get_table_names()

if "question_answer" in tables:
    print("Table exists.")
else:
    print("Table does not exist.")
    
# Print all table names
print(tables)
```

# Check the content of created table

```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import create_engine, Column, Integer, Text, Table, MetaData
from dotenv import load_dotenv, find_dotenv
import os

load_dotenv(find_dotenv('../application/.env'))

# Retrieve the environment variables for the connection
host = os.getenv("PG_VECTOR_HOST")
user = os.getenv("PG_VECTOR_USER")
password = os.getenv("PG_VECTOR_PASSWORD")
database = os.getenv("PGDATABASE5")  # question_answer

# Create the connection string
CONNECTION_STRING = f"postgresql+psycopg://{user}:{password}@{host}:5432/{database}"
print(CONNECTION_STRING)

# Create a base class using declarative_base
Base = declarative_base()

# Define your table as a model class
class FAQ(Base):
    __tablename__ = 'question_answer'  # Table name in the database

    id = Column(Integer, primary_key=True, autoincrement=True)  # Auto-incrementing primary key
    question = Column(Text, nullable=False)  # Question column of type text
    answer = Column(Text, nullable=False)  # Answer column of type text
    bot_type = Column(Text, nullable=False)  # Bot type column of type text
    
# Create an engine that connects to the PostgreSQL database
engine = create_engine(CONNECTION_STRING)

# Create a configured "Session" class
Session = sessionmaker(bind=engine)

# Create a session
session = Session()

# Query all records from the FAQ table
faqs = session.query(FAQ).all()

# Loop through the results and print them
for faq in faqs:
    print(f"ID: {faq.id}, Question: {faq.question}, Answer: {faq.answer}, Bot Type: {faq.bot_type}")

```

# Insert a new row

```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker

# Create an engine that connects to the PostgreSQL database
engine = create_engine(CONNECTION_STRING)

# Create a configured "Session" class
Session = sessionmaker(bind=engine)

# Create a session
session = Session()

# Example data
question = "What is AI?"
answer = "AI stands for Artificial Intelligence."
bot_type = "Educational"

# Create a new FAQ instance
new_faq = FAQ(question=question, answer=answer, bot_type=bot_type)

# Add the new instance to the session
session.add(new_faq)

# Commit the session to insert the data into the table
session.commit()

```

# Drop all rows from the table

```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker

# Create an engine that connects to the PostgreSQL database
engine = create_engine(CONNECTION_STRING)

# Create a configured "Session" class
Session = sessionmaker(bind=engine)

# Create a session
session = Session()

# Delete all rows in the FAQ table
session.query(FAQ).delete()

# Commit the transaction to apply the changes
session.commit()

# Close the session
session.close()
```
